\begin{equation}
\frac{dx}{dt} = \sigma(y-x)
\end{equation}

\begin{equation}
\frac{dy}{dt} = x(\rho-z)-y
\end{equation}

\begin{equation}
\frac{dz}{dt} = xy-\beta z
\end{equation}

where $\sigma,\rho, \beta $ are constants.




```python
from ipywidgets import*  ##importing librarie
import numpy as np
import ipyvolume as ipv ##this is the best library for 3d plotting
from IPython.display import display ##this is for rendering javascripts



def lorenz(x, y, z, s,r,b):   ##defining the lorenz function
    '''
    Given:
       x, y, z: a point of interest in three dimensional space
       sig, rho, beta: parameters defining the lorenz attractor
    Returns:
       x_dot, y_dot, z_dot: values of the lorenz attractor's partial
           derivatives at the point x, y, z
    '''
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot


dt = 0.01  ##time step


##creating figure and one plot1 object  whose graph element will be changed in the main function

fig = ipv.figure()
plot1 = ipv.plot(np.array([0.]),np.array([0.]),np.array([0.]))
ipv.show()
################################################333

#############################
###Creating Main function that will communicate with sliders and will also plot x,y,z in 3D
def f_main(sig,rho,beta,x_0,y_0,z_0,N_iter,plot):
    global plot1,fig  ##because we have already defined them above that's why declaring them as global
    if(plot==True):  ##function will start integrating only when plot button is checked
        xs = np.array([x_0])  ## x_0,y_0,z_0 are initial conditions
        ys = np.array([y_0])
        zs = np.array([z_0])

        for i in range(N_iter): ##N_iter is iteration number it can be changed
            x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i],sig,rho,beta) ##calculating dx/dt,dy/dt,dz/dt
            ###########updating x,y,z
            x_new = xs[i] + (x_dot * dt) 
            y_new = ys[i] + (y_dot * dt)
            z_new = zs[i] + (z_dot * dt)
            ###updating array 
            xs = np.append(xs,x_new)
            ys = np.append(ys,y_new)
            zs = np.append(zs,z_new)
        ########################333
        
        ######3plotting x,y,z, it just requires to change x,y,z array of plot1 object
        plot1.x = xs
        fig.xlim = (-1*np.max(xs),np.max(xs))
        plot1.y = zs
        fig.zlim = (-1*np.max(ys),np.max(ys))
        plot1.z = ys
        fig.ylim = (0,np.max(zs))       

        
                
print("Uncheck plot button  then change the sliders") ##just wanna give a warning
        

    
########################
#Creating interactive function that establishes communication between sliders and f_main
w = interactive(f_main, sig = FloatSlider(value = 10,min = 0.1,max = 50.0,layout ={'width':'700px'}),
                    rho = FloatSlider(value = 28,min = 0.1,max = 50.0,layout ={'width':'700px'}),
                    beta = FloatSlider(value = 2.667,min = 0.1,max = 50.0,layout ={'width':'700px'}),
                    x_0 = FloatSlider(value = 0,min = -5.00,max = 5.00,step = 0.01,layout ={'width':'700px'}),
                     y_0 = FloatSlider(value = 1.00,min = -5.00,max = 5.00,step = 0.01,layout ={'width':'700px'}),
                     z_0 = FloatSlider(value = 1.07,min = -5.00,max = 5.00,step = 0.01,layout ={'width':'700px'}),
                    N_iter= IntSlider(description="Iteartion ",value = 5000,min = 1000,step = 1000,max = 50000.0,layout ={'width':'700px'}),
                    plot= Checkbox(value = True,description=" Plot Curve")
                    
            )
display(w)
```
